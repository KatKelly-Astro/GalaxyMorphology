```python
import matplotlib.pyplot as plt
fig, ax=plt.subplots()
ax.hist(f['Mass'], bins=[500000, 600000, 700000, 800000])
ax.set_xticks([500000, 600000, 700000, 800000])
plt.xlabel('Mass ($M_{\odot}$)')
plt.ylabel("Count")
plt.title('Mass Histogram at redshift = 7.050')
plt.show()
```

```python
fig, ax=plt.subplots()

Co = (f['Coordinates'])
x = Co[:,0]
y = Co[:,1]
ax.scatter(x, y)
#plt.rcParams["figure.figsize"] = (10, 10)
plt.xlabel("X coordinates (Mpc)")
plt.ylabel("Y coordinates (Mpc)")
plt.ylim(-0.002,0.004)
plt.xlim(-0.003,0.003)
plt.title('X vs Y Coordinates at redshift = 7.050')
plt.show()
```
