```python
import pandas as pd
from pandas import DataFrame
import csv
df_organic = pd.read_csv("halo_catalogue_organic_new.txt", delim_whitespace=True, header=None, skiprows=(0, 0), names=['Index','Time (Gyr)','a_exp','magnitude','subhalo centre X','subhalo centre Y','subhalo centre Z','subhalo peculiar velocity X','subhalo peculiar velocity Y','subhalo peculiar velocity Z','Halo mass','Stellar mass','BH mass','SFR','sSFR'] )
df_early = pd.read_csv("halo_catalogue_gm_early_new.txt", delim_whitespace=True, header=None, skiprows=(0, 0), names=['Index','Time (Gyr)','a_exp','magnitude','subhalo centre X','subhalo centre Y','subhalo centre Z','subhalo peculiar velocity X','subhalo peculiar velocity Y','subhalo peculiar velocity Z','Halo mass','Stellar mass','BH mass','SFR','sSFR'] )
df_late = pd.read_csv("halo_catalogue_gm_late_new.txt", delim_whitespace=True, header=None, skiprows=(0, 0), names=['Index','Time (Gyr)','a_exp','magnitude','subhalo centre X','subhalo centre Y','subhalo centre Z','subhalo peculiar velocity X','subhalo peculiar velocity Y','subhalo peculiar velocity Z','Halo mass','Stellar mass','BH mass','SFR','sSFR'] )
df_organic
```

```python
# Halo Mass plots
# To do:
# deviation


import matplotlib.pyplot as plt
# import numpy as np

# Create a data frame for each of the catalogues
plt.plot(df_early['Time (Gyr)'],df_early['Halo mass'], linewidth=1, color='maroon', label='Early')

plt.plot(df_late['Time (Gyr)'],df_late['Halo mass'], linewidth=1, color='blue', label='Late')

plt.plot(df_organic['Time (Gyr)'],df_organic['Halo mass'], linewidth=1, color='darkgreen', label='Organic')

# Plot the y-axis on a log scale
plt.yscale("log")

# Label the x and y axis
plt.xlabel('Time (Gyr)')
plt.ylabel('$Log_{10}$ Halo Mass ($M_{\odot}$)')

# Add a legend using the labels in the df plot lines above
plt.legend(loc="lower right")

# Add a plot title
plt.title('Halo mass variation')

# Show plot
plt.show

# OPTIONAL FORMATTING
#ax = plt.axes()
#ax.set_facecolor("grey")
#plt.grid(color='red', linestyle='-.', linewidth=0.7)
#x_ticks = np.arange(0, 15, 1)
#plt.xticks(x_ticks)
```
