```python
import pandas as pd
import csv
df_organic = pd.read_csv("halo_catalogue_organic.txt", delim_whitespace=True, header=None, skiprows=(0, 0), names=['Index','Time (Gyr)','a_exp','magnitude','subhalo centre X','subhalo centre Y','subhalo centre Z','subhalo peculiar velocity X','subhalo peculiar velocity Y','subhalo peculiar velocity Z','Halo mass','Stellar mass','BH mass','SFR','sSFR'] )
df_early = pd.read_csv("halo_catalogue_gm_early.txt", delim_whitespace=True, header=None, skiprows=(0, 0), names=['Index','Time (Gyr)','a_exp','magnitude','subhalo centre X','subhalo centre Y','subhalo centre Z','subhalo peculiar velocity X','subhalo peculiar velocity Y','subhalo peculiar velocity Z','Halo mass','Stellar mass','BH mass','SFR','sSFR'] )
df_late = pd.read_csv("halo_catalogue_gm_late.txt", delim_whitespace=True, header=None, skiprows=(0, 0), names=['Index','Time (Gyr)','a_exp','magnitude','subhalo centre X','subhalo centre Y','subhalo centre Z','subhalo peculiar velocity X','subhalo peculiar velocity Y','subhalo peculiar velocity Z','Halo mass','Stellar mass','BH mass','SFR','sSFR'] )
```

```python
# Halo Mass plots
# To do:
# All on one graph
# y axis ticks
# deviation
# subscripts in axis titles?

import matplotlib.pyplot as plt
import numpy as np

df_early.plot(x='Time (Gyr)',y='Halo mass', linewidth=2, color='maroon')
plt.yscale("log")
plt.xlabel('Time (Gyr)')
plt.ylabel('log10 Halo Mass (M_sun)')
L=plt.legend()
x_ticks = np.arange(0, 15, 1)
plt.xticks(x_ticks)
L.get_texts()[0].set_text('Early')

df_late.plot(x='Time (Gyr)',y='Halo mass', linewidth=2, color='blue')
plt.yscale("log")
plt.xlabel('Time (Gyr)')
plt.ylabel('log10 Halo Mass (M_sun)')
L=plt.legend()
x_ticks = np.arange(0, 15, 1)
plt.xticks(x_ticks)
L.get_texts()[0].set_text('Late')

df_organic.plot(x='Time (Gyr)',y='Halo mass', linewidth=2, color='darkgreen')
plt.yscale("log")
plt.xlabel('Time (Gyr)')
plt.ylabel('log10 Halo Mass (M_sun)')
L=plt.legend()
x_ticks = np.arange(0, 15, 1)
plt.xticks(x_ticks)
L.get_texts()[0].set_text('Organic')
```
